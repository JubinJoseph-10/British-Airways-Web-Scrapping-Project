```python
#Importing Libraries
from bs4 import BeautifulSoup
import requests,openpyxl
import pandas as pd
import numpy as np
```

```python
#Creating a variable with the source code of the website and checing it
#Putting the same in a try and except block for errors

try:
    source = requests.get('https://www.airlinequality.com/airline-reviews/british-airways/?sortby=post_date%3ADesc&pagesize=3537')
    source.raise_for_status()
    
    soup=BeautifulSoup(source.text,'html.parser')
    ratings = soup.find('article', class_='comp comp_reviews-airline querylist position-content').find_all('article')
     

except Exception as e:
    print(e)
```

```python
#Generating Basic Details  
#Creating empty lists where data would be plugged in

rev_title=[]
rev_ovr=[]
cust_name=[]
cust_country=[]
rev_date=[]
cust_verified=[]

#Using loops and try and except blocks to exract data
#each time the for loop runs it appends a new value to the lists above

for rating in ratings:
        
        try:
            review_title = rating.find('h2',class_='text_header').text
            rev_title.append(str(review_title))
        except:
            review_title = np.nan
            rev_title.append(review_title)
        
        try:
            ovr_review = rating.find('div',class_='rating-10').span.text
            rev_ovr.append(ovr_review)
        except:
            ovr_review = np.nan
            rev_ovr.append(ovr_review)
            
        try:  
            name = rating.find('span',itemprop='name').text
            cust_name.append(str(name))
        except:
            name = np.nan
            cust_name.append(name)
            
        try:
            txt = rating.find('h3',class_='text_sub_header userStatusWrapper').text
            start = txt.index('(')
            end = txt.index(')')
            country = txt[start+1:end]
            cust_country.append(country)
    
        except:
            country=np.nan
            cust_country.append(country)
            
        
        try:
            review_date=rating.find('time', itemprop='datePublished').text
            rev_date.append(str(review_date))
        except:
            review_date=np.nan
            rev_date.append(review_date)
            
        try:
            trip_ver = rating.find('em').text
            cust_verified.append(str(trip_ver))
        except:
            trip_ver= np.nan
            cust_verified.append(trip_ver)
            

#putting together all the new lists into a newly created dataframe            
cat1_data_British_Airways=pd.DataFrame(list(zip(cust_name,cust_country,rev_date,rev_title,rev_ovr,cust_verified)),
                                      columns=['cust_name','cust_country','rev_date','rev_title','rev_ovr', 'cust_verified'])

cat1_data_British_Airways['rev_ovr']=pd.to_numeric(cat1_data_British_Airways['rev_ovr'])
```

```python
#Creation of an empty list 
aircraft_name=[]

#Getting the AirCraft Names from the website
for rating in ratings:
    tester = str(rating.find(string='Aircraft'))
    if tester == 'Aircraft':
        dell=[]
        rows = rating.find_all('td')
        for row in rows:
            dell.append(str(row))
        c = dell.index('<td class="review-rating-header aircraft">Aircraft</td>')
        Air_Name=rows[c+1].text
        aircraft_name.append(str(Air_Name))
    else:
        Air_Name=np.nan
        aircraft_name.append(Air_Name)
```

```python
# Creating an empty list
traveller_type=[]

# Getting the Type of Traveller from the website
for rating in ratings:
    tester = str(rating.find(string='Type Of Traveller'))
    if tester == 'Type Of Traveller':
        dell=[]
        rows = rating.find_all('td')
        for row in rows:
            dell.append(str(row))
        c = dell.index('<td class="review-rating-header type_of_traveller">Type Of Traveller</td>')
        Traveller_Type=rows[c+1].text
        traveller_type.append(str(Traveller_Type))
    else:
        Traveller_Type=np.nan
        traveller_type.append(Traveller_Type)
```

```python
#creating an empty list
seat_type=[]

# Getting the Seat Type from the website
for rating in ratings:
    tester = str(rating.find(string='Seat Type'))
    if tester == 'Seat Type':
        dell=[]
        rows = rating.find_all('td')
        for row in rows:
            dell.append(str(row))
        c = dell.index('<td class="review-rating-header cabin_flown">Seat Type</td>')
        Seat_type=rows[c+1].text
        seat_type.append(str(Seat_type))
    else:
        Seat_type=np.nan
        seat_type.append(Seat_type)
```

```python
#Creating an empty list
route_=[]

# Getting the Route from the website
for rating in ratings:
    tester = str(rating.find(string='Route'))
    if tester == 'Route':
        dell=[]
        rows = rating.find_all('td')
        for row in rows:
            dell.append(str(row))
        c = dell.index('<td class="review-rating-header route">Route</td>')
        route=rows[c+1].text
        route_.append(str(route))
    else:
        Seat_type=np.nan
        route_.append(route)
```

```python
#creating an empty list
fly_date=[]

# Getting the Flying Date from the website
for rating in ratings:
    tester = str(rating.find(string='Date Flown'))
    if tester == 'Date Flown':
        dell=[]
        rows = rating.find_all('td')
        for row in rows:
            dell.append(str(row))
        c = dell.index('<td class="review-rating-header date_flown">Date Flown</td>')
        flying_date=rows[c+1].text
        fly_date.append(str(flying_date))
    else:
        flying_date=np.nan
        fly_date.append(flying_date)
```

```python
#Putting the remaining categorival variables in the list
cat2_data_British_Airways=pd.DataFrame(list(zip(aircraft_name,traveller_type,seat_type,route_,fly_date)),
                                      columns=['aircraft_name','traveller_type','seat_type','route_','fly_date'])
```

```python
# concatenating cat2_data_British_Airways and cat1_data_British_Airways along columns
cat_data_British_Airways = pd.concat([cat1_data_British_Airways,cat2_data_British_Airways ], axis=1)
```
